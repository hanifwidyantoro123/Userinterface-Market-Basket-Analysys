PENGUJIAN SUPPORT & CONFIDENCE TERGADAP RULE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Membaca data dari file excel
df = pd.read_excel('Dataset Transaksi Medicom Sungkono 2021_2022.xlsx')
```

```python
# Memilih kolom yang diperlukan dan mengubah nama kolom 'id customer' menjadi 'id transaksi'
selected_column = ['Date', 'ID CUSTOMER','NAMA BARANG','HARGA SATUAN','SUBTOTAL']
df = df[selected_column]
df.rename(columns={'ID CUSTOMER':'ID TRANSAKSI'},inplace=True)
```

```python
#menambahkan kolom QTY dari subtotal/harga satuan
df['QTY'] = df['SUBTOTAL']/df['HARGA SATUAN']
```

```python
#Membersihkan data duplikat
df_duplicate = df.drop_duplicates(keep=False)
df = df_duplicate
```

```python
# Pastikan kolom 'date' dalam format datetime 
# Mengonversi kolom date ke format datetime untuk memudahkan pemisahan berdasarkan tahun.
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.shape
```

```python
#Data Transformation, Mengubah menjadi Data Tabular
data_pivot= df.pivot_table(
        index='ID TRANSAKSI', columns='NAMA BARANG', values='QTY', aggfunc='sum'
    ).fillna(0)
data_pivot 
```

```python
#code untuk merubah 0 dan 1 pada tabel tabular
def encode(x):
    """Encodes values as 0 (if <= 0) or 1 (if >= 1)."""
    if x <= 0:
        return 0
    elif x >= 1:
        return 1
```

```python
#Merubah menjadi integer
data_pivot = data_pivot .astype("int64")
```

```python
#merubah integer menjadi bilangan biner 0 dan 1
data_pivot = data_pivot .applymap(encode)
data_pivot
```

```python
#Minimum Support dan Frequent Itemset
support = 0.005
frequent_items= apriori(data_pivot, min_support = round(support, 3), use_colnames=True)
# frequent_items.sort_values("support", ascending=False)
# Mengubah nilai support menjadi persen
# frequent_items['support_percent'] = frequent_items['support'] * 9904

# Menambahkan kolom panjang itemset
# frequent_items['length'] = frequent_items['itemsets'].apply(lambda x: len(x))

# Filter untuk hanya menampilkan itemset dengan panjang 2 (2-itemsets)
# frequent_items_2_itemsets = frequent_items[frequent_items['length'] == 3]

# Mengurutkan berdasarkan nilai support yang telah diubah menjadi persen
# frequent_items_2_itemsets.sort_values("support_percent", ascending=False, inplace=True)

# Menampilkan hasil
# frequent_items
```

```python
#Confidence & Rule
metric = "lift"
min_threshold = 0.5

rules= association_rules(frequent_items, metric = metric, min_threshold= min_threshold)[["antecedents", "consequents","support","confidence","lift"]]
# rules.sort_values(by=['confidence'], ascending = False)
filtered_rules = rules.sort_values('confidence', ascending=False)
filtered_rules
```

```python
filtered_rules.to_excel('association-rule(0.007).xlsx')
```

```python
filtered_rules.shape
```

```python

# Mengurutkan berdasarkan confidence
filtered_rules = rules.sort_values('confidence', ascending=False)

# Menghitung frekuensi A (jumlah transaksi yang mengandung antecedent)
def get_frequency_A(row, data_pivot):
    antecedent_items = list(row['antecedents'])
    return data_pivot[antecedent_items].all(axis=1).sum()

filtered_rules['frekuensi_A'] = filtered_rules.apply(lambda row: get_frequency_A(row, data_pivot), axis=1)

# Menghitung frekuensi A ∩ B (jumlah transaksi yang mengandung antecedent dan consequent)
def get_frequency_A_n_B(row, data_pivot):
    antecedent_items = list(row['antecedents'])
    consequent_items = list(row['consequents'])
    return data_pivot[antecedent_items + consequent_items].all(axis=1).sum()

filtered_rules['frekuensi_A_n_B'] = filtered_rules.apply(lambda row: get_frequency_A_n_B(row, data_pivot), axis=1)

# Mengonversi confidence menjadi bentuk persentase dan menambahkan penjelasan
filtered_rules['confidence_str'] = filtered_rules.apply(
    lambda row: f"({int(row['frekuensi_A_n_B'])}/{int(row['frekuensi_A'])})={row['confidence']:.2f} ({row['confidence'] * 100:.0f}%)",
    axis=1
)

# Mengubah kolom antecedents dan consequents menjadi string
filtered_rules['antecedents'] = filtered_rules['antecedents'].apply(lambda x: ', '.join(list(x)))
filtered_rules['consequents'] = filtered_rules['consequents'].apply(lambda x: ', '.join(list(x)))

# Menyusun data dalam format tabel yang diinginkan
result = filtered_rules[['antecedents', 'frekuensi_A', 'frekuensi_A_n_B', 'lift', 'confidence_str']]
result.columns = ['Pola 2 Item', 'Frekuensi A', 'Frekuensi A ∩ B', 'Lift', 'Confidence']

# Menampilkan hasil
result
```

```python
#code untuk merubah 0 dan 1 pada tabel tabular
def encode(x):
    """Encodes values as 0 (if <= 0) or 1 (if >= 1)."""
    if x <= 0:
        return 0
    elif x >= 1:
        return 1
```

```python
support = 0.005
frequent_items_date= apriori(data_pivot_date, min_support = round(support, 3), use_colnames=True)
frequent_items_date.sort_values("support", ascending=False)
```

```python
def calculate_total_quantity(itemsets, df):
    total_quantities = []
    for itemset in itemsets:
        itemset_list = list(itemset)
        subset = df[df[itemset_list].sum(axis=1) == len(itemset_list)]
        total_quantity = subset[itemset_list].sum().sum()
        total_quantities.append(total_quantity)
    return total_quantities

# Menghitung total quantity untuk setiap itemset
frequent_items_date['total_quantity'] = calculate_total_quantity(frequent_items_date['itemsets'], data_pivot_date)
frequent_items_date.sort_values("support", ascending=False)
```

```python
frequent_items_date.to_excel('totalFrequent-Items.xlsx')
```

```python
itemsets = {}
max_length = frequent_items_date['itemsets'].apply(len).max()

for i in range(1, max_length + 1):
    itemsets[i] = frequent_items_date[frequent_items_date['itemsets'].apply(len) == i]

# Menampilkan itemsets berdasarkan panjangnya
# for length, filtered_df_date in itemsets.items():
#     print(f"Itemsets dengan panjang {length}:\n", filtered_df_date)

# Menyimpan itemsets ke dalam file Excel
with pd.ExcelWriter('frequent_itemsets_3bln dan qty.xlsx') as writer:
    for length, filtered_df_date in itemsets.items():
        filtered_df_date.to_excel(writer, sheet_name=f'{length}-itemsets', index=False)

print("Frequent itemsets telah disimpan ke dalam file 'frequent_itemsets_3bln dan qty.xlsx'")
```
