```python
import pandas as pd
from sklearn.model_selection import KFold
from mlxtend.frequent_patterns import apriori, association_rules

# Membaca data transaksi dari file CSV
# data = pd.read_csv('transactions.csv')
# Untuk demonstrasi, kita buat data dummy
data = pd.DataFrame({
    'TransactionID': [1, 1, 2, 2, 3],
    'CustomerID': [101, 101, 102, 102, 103],
    'Item': ['Milk', 'Bread', 'Butter', 'Milk', 'Bread']
})

# Melakukan pivot untuk mendapatkan data dalam bentuk one-hot encoding
basket = (data.groupby(['TransactionID', 'Item'])['Item']
          .count().unstack().reset_index().fillna(0)
          .set_index('TransactionID'))
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

# Definisi fungsi cross-validation
def cross_validation(data, n_splits, min_support_values, min_confidence_values):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    results = []

    for train_index, test_index in kf.split(data):
        train_data, test_data = data.iloc[train_index], data.iloc[test_index]

        for min_support in min_support_values:
            for min_confidence in min_confidence_values:
                # Generate frequent itemsets
                frequent_itemsets = apriori(train_data, min_support=min_support, use_colnames=True)
                # Generate association rules
                rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
                # Evaluate rules
                score = evaluate_rules(rules, test_data)  # Implement this function based on your metrics
                results.append((min_support, min_confidence, score))

    return results

# Definisi fungsi evaluasi (contoh menggunakan lift)
def evaluate_rules(rules, test_data):
    # Contoh sederhana: menjumlahkan lift dari semua aturan
    if rules.empty:
        return 0
    return rules['lift'].sum()

# Parameter untuk cross-validation
n_splits = 3
min_support_values = [0.1, 0.2]
min_confidence_values = [0.5, 0.7]

# Menjalankan cross-validation
results = cross_validation(basket, n_splits, min_support_values, min_confidence_values)

# Menemukan parameter terbaik
best_result = max(results, key=lambda x: x[2])
best_support = best_result[0]
best_confidence = best_result[1]
best_score = best_result[2]

print(f"Best support: {best_support}, Best confidence: {best_confidence}, Best score: {best_score}")
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.model_selection import train_test_split

# Membaca data transaksi dari file CSV
# data = pd.read_csv('transactions.csv')
# Untuk demonstrasi, kita buat data dummy


# Split data into training and testing


# Generate association rules
min_support = 0.002
min_confidence = 0.03

frequent_itemsets = apriori(data_training_pivot, min_support=min_support, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

# Evaluate rules on test data
def evaluate_rules(rules, test_data):
    if rules.empty:
        return []
    
    test_data['transaction'] = test_data.index
    test_data = test_data.melt(id_vars='transaction', var_name='item', value_name='presence')
    test_data = test_data[test_data['presence'] == 1]

    def rule_applies(rule, transaction_items):
        return all(item in transaction_items for item in rule['antecedents']) and \
               all(item in transaction_items for item in rule['consequents'])

    detailed_results = []
    for _, rule in rules.iterrows():
        true_positives = 0
        false_positives = 0
        false_negatives = 0
        for _, transaction in test_data.groupby('transaction'):
            transaction_items = set(transaction['item'])
            if rule_applies(rule, transaction_items):
                true_positives += 1
            else:
                if all(item in transaction_items for item in rule['antecedents']):
                    false_positives += 1
                if all(item in transaction_items for item in rule['consequents']):
                    false_negatives += 1
        
        precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
        recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        accuracy = true_positives / len(test_data) if len(test_data) > 0 else 0

        if accuracy > 0:
            detailed_results.append({
                'antecedents': rule['antecedents'],
                'consequents': rule['consequents'],
                'support': rule['support'],
                'confidence': rule['confidence'],
                'lift': rule['lift'],
                'precision': precision,
                'recall': recall,
                'f1_score': f1_score,
                'accuracy': accuracy
            })

    return detailed_results

detailed_results = evaluate_rules(rules, data_testing_pivot)

# Membuat DataFrame dari detailed_results
df_results = pd.DataFrame(detailed_results)

# Mengurutkan berdasarkan accuracy secara descending
df_results_sorted = df_results.sort_values(by='accuracy', ascending=False).reset_index(drop=True)

# Menampilkan tabel hasil
print(df_results_sorted)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Contoh data untuk demonstrasi
# Data tahun 2021 (Training Data)
data_2021 = pd.DataFrame({
    'TransactionID': [1, 1, 2, 2, 3],
    'CustomerID': [101, 101, 102, 102, 103],
    'Item': ['Milk', 'Bread', 'Butter', 'Milk', 'Bread']
})

# Data tahun 2022 (Testing Data)
data_2022 = pd.DataFrame({
    'TransactionID': [4, 4, 5, 6, 6],
    'CustomerID': [104, 104, 105, 106, 106],
    'Item': ['Butter', 'Milk', 'Bread', 'Milk', 'Butter']
})

# Melakukan pivot untuk mendapatkan data dalam bentuk one-hot encoding untuk data training
basket_2021 = (data_2021.groupby(['TransactionID', 'Item'])['Item']
               .count().unstack().reset_index().fillna(0)
               .set_index('TransactionID'))
basket_2021 = basket_2021.applymap(lambda x: 1 if x > 0 else 0)

# Melakukan pivot untuk mendapatkan data dalam bentuk one-hot encoding untuk data testing
basket_2022 = (data_2022.groupby(['TransactionID', 'Item'])['Item']
               .count().unstack().reset_index().fillna(0)
               .set_index('TransactionID'))
basket_2022 = basket_2022.applymap(lambda x: 1 if x > 0 else 0)

# Fungsi untuk evaluasi (contoh menggunakan confidence)
def evaluate_rules(rules, test_data):
    if rules.empty:
        return 0, 0, 0, [], []

    matched_rules = 0
    total_rules = len(rules)
    
    matching_rules = []
    non_matching_rules = []
    
    for _, rule in rules.iterrows():
        antecedents = set(rule['antecedents'])
        consequents = set(rule['consequents'])

        rule_match = False
        for _, transaction in test_data.iterrows():
            transaction_items = set(transaction[transaction == 1].index)
            if antecedents.issubset(transaction_items) and consequents.issubset(transaction_items):
                matched_rules += 1
                rule_match = True
                break
        
        if rule_match:
            matching_rules.append(rule)
        else:
            non_matching_rules.append(rule)

    precision = matched_rules / total_rules
    recall = matched_rules / len(test_data)
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return precision, recall, f1_score, matching_rules, non_matching_rules

# Fungsi untuk melatih dan mengevaluasi model
def train_and_evaluate(train_data, test_data, min_support, min_confidence):
    # Generate frequent itemsets
    frequent_itemsets = apriori(data_training_pivot, min_support=min_support, use_colnames=True)
    if frequent_itemsets.empty:
        return None, 0, 0, 0, [], []

    # Generate association rules
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
    if rules.empty:
        return None, 0, 0, 0, [], []

    # Evaluate rules
    precision, recall, f1_score, matching_rules, non_matching_rules = evaluate_rules(rules, data_testing_pivot)
    return rules, precision, recall, f1_score, matching_rules, non_matching_rules

# Menentukan nilai support dan confidence
min_support_values = [0.002,0.03]
min_confidence_values = [0.3,0.5]

# Menjalankan pelatihan dan evaluasi untuk berbagai nilai support dan confidence
results = []

for min_support in min_support_values:
    for min_confidence in min_confidence_values:
        rules, precision, recall, f1_score, matching_rules, non_matching_rules = train_and_evaluate(basket_2021, basket_2022, min_support, min_confidence)
        results.append((min_support, min_confidence, precision, recall, f1_score, matching_rules, non_matching_rules))

# Membuat DataFrame untuk hasil
results_df = pd.DataFrame([(res[0], res[1], res[2], res[3], res[4]) for res in results], 
                          columns=['min_support', 'min_confidence', 'precision', 'recall', 'f1_score'])

# Menemukan parameter terbaik
best_result = results_df.loc[results_df['f1_score'].idxmax()]
best_support = best_result['min_support']
best_confidence = best_result['min_confidence']
best_f1_score = best_result['f1_score']

print(f"Best support: {best_support}, Best confidence: {best_confidence}, Best F1 Score: {best_f1_score}")
print("\nResults Table:")
print(results_df)

# Menampilkan aturan yang memenuhi dan tidak memenuhi
best_matching_rules = results[results_df['f1_score'].idxmax()][5]
best_non_matching_rules = results[results_df['f1_score'].idxmax()][6]

print("\nMatching Rules:")
print(pd.DataFrame(best_matching_rules))

print("\nNon-Matching Rules:")
print(pd.DataFrame(best_non_matching_rules))
```
